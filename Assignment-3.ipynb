# Using Replace Function

```python
import pandas as pd
df=pd.read_csv('insurance.csv')
df.head()
```

```python
df.sex.unique()
```

```python
df.sex=df.sex.replace(['female','male'],[0,1])
df.head()
```

```python
df.smoker.unique()
```

```python
df.smoker=df.smoker.replace(['yes','no'],[0,1])
df.head()
```

```python
df.region.unique()
```

```python
df.region=df.region.replace(['southwest','southeast','northwest','northeast'],[0,1,2,3])
df.head()
```

# Label Encoding

```python
import pandas as pd
df=pd.read_csv('insurance.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in df.columns:
    if is_numeric_dtype(df[col]):
        continue
    else:
        df[col]=le.fit_transform(df[col])
```

```python
df.head()
```

# One-Hot-Encoding Using Loop

```python
import pandas as pd
df=pd.read_csv('insurance.csv')
df.head()
```

```python
columns=['sex','smoker','region']
for col in columns:
    one_hot=pd.get_dummies(df[col],drop_first=True)
    df=pd.concat((df,one_hot),axis=1).drop(col,axis=1)
```

```python
df.head()
```

```python
y=df['charges']
y.head()
```

```python
x=df.drop('charges',axis=1)
x.head()
```

# Ordinal Encoder

```python
import pandas as pd
df=pd.read_csv('insurance.csv')
df.head()
```

```python
df.sex.unique()
```

```python
sex=['female','male']
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[sex])
```

```python
encoded=oe.fit_transform(df[['sex']])
```

```python
df1=pd.DataFrame(encoded,columns=['Sex'])
df1.head()
```

```python
df.sex=df1.Sex
```

# Ordinal Encoding Using Loop

```python
import pandas as pd
df=pd.read_csv('insurance.csv')
df.head()
```

```python
cols=['sex','smoker','region']
for col in cols:
    unq=df[col].unique()
    df[col]=OrdinalEncoder(categories=[unq]).fit_transform(df[[col]])
```
